```python
# imports
import subprocess
import json
import re
from typing import Literal
import os


def fix_metadata(name: str, to_fix: Literal["X", "Y"], series: int):
    # generate inputs
    to_fix_idx = 4 if to_fix == "X" else 3
    other = "Y" if to_fix == "X" else "X"
    other_idx = 3 if to_fix == "X" else 4

    # download original metadata
    cmd = [
        "mc",
        "cp",
        f"culcol_s3_rw/culture-collections/data/single_volumes/images/ome-zarr/{name}.ome.zarr/OME/METADATA.ome.xml",
        "METADATA_original.ome.xml",
    ]
    subprocess.run(cmd)

    # load original metadata
    with open("METADATA_original.ome.xml", "r") as f:
        xml_string = f.read()

    # find value to use
    pattern = re.compile(rf'PhysicalSize{other}="(\d+\.\d+)"')
    match = pattern.search(xml_string)
    value = match.group(1)

    # replace value in xml_string
    new_xml = re.sub(
        rf'PhysicalSize{to_fix}="(\d+\.\d+)"',
        rf'PhysicalSize{to_fix}="{value}"',
        xml_string,
    )

    # write new xml_string to file
    with open("METADATA.ome.xml", "w") as f:
        f.write(new_xml)

    # upload new metadata
    cmd = [
        "mc",
        "cp",
        "METADATA.ome.xml",
        f"culcol_s3_rw/culture-collections/data/single_volumes/images/ome-zarr/{name}.ome.zarr/OME/METADATA.ome.xml",
    ]
    subprocess.run(cmd)

    # download zattrs
    cmd = [
        "mc",
        "cp",
        f"culcol_s3_rw/culture-collections/data/single_volumes/images/ome-zarr/{name}.ome.zarr/{series}/.zattrs",
        ".zattrs_original",
    ]
    subprocess.run(cmd)

    # load original zattrs
    with open(".zattrs_original", "r") as f:
        zattrs = json.load(f)

    # replace value in zattrs
    for vol in zattrs["multiscales"][0]["datasets"]:
        scale = vol["coordinateTransformations"][0]["scale"]
        vol["coordinateTransformations"][0]["scale"] = (
            scale[:to_fix_idx] + [scale[other_idx]] + scale[to_fix_idx + 1 :]
        )

    # write new zattrs to file
    with open(".zattrs", "w") as f:
        json.dump(zattrs, f, indent=2)

    # upload new zattrs
    cmd = [
        "mc",
        "cp",
        ".zattrs",
        f"culcol_s3_rw/culture-collections/data/single_volumes/images/ome-zarr/{name}.ome.zarr/{series}/.zattrs",
    ]
    subprocess.run(cmd)

    # remove temporary files
    os.remove("METADATA_original.ome.xml")
    os.remove("METADATA.ome.xml")
    os.remove(".zattrs_original")
    os.remove(".zattrs")
```

```python
fix_metadata(
    name="20240401_ZE10_fm_DNA_Centr_tub_NHS_60x4x_005", to_fix="X", series=0
)
fix_metadata(
    name="20240401_ZE10_fm_DNA_Centr_tub_NHS_60x4x_006", to_fix="X", series=0
)
fix_metadata(
    name="20240401_ZE10_fm_DNA_Centr_tub_NHS_60x4x_007", to_fix="X", series=0
)
fix_metadata(
    name="20230621_Tokyo_chem_NHS_tub_centr_DNA_021", to_fix="Y", series=0
)
```
