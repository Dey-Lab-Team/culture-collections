```python
import pandas as pd

df = pd.read_csv("name_map_iter3_in.csv", sep="\t")
df
```

```python
new_names_iter3 = []
double_check = [False] * len(df)
for i, row in df.iterrows():
    new_name_iter2 = row["new_filename_iteration2"]
    new_channels = str(row["relabel_channels_clean"])
    if row.isna().any():
        new_names_iter3.append(new_name_iter2)
        print(f"{new_name_iter2}\t{double_check[i]}")
        continue
    if new_channels == "TODO" or new_channels == "ALSO?":
        double_check[i] = True
        new_names_iter3.append(new_name_iter2)
        print(f"{new_name_iter2}\t{double_check[i]}")
        continue
    name_split = new_name_iter2.split("_")
    name_split[2] = new_channels
    new_name_iter3 = "_".join(name_split)
    new_names_iter3.append(new_name_iter3)
    print(f"{new_name_iter3}\t{double_check[i]}")
df["new_name_iter3"] = new_names_iter3
df["double_check"] = double_check
df.to_csv("name_map_iter3_out.csv", sep="\t", index=False)
```
