```python
import numpy as np
import pandas as pd
```

```python
table = pd.read_csv('final_name_map.csv', sep='\t')
table
```

```python
ids = set()
fixes = set()
channels = set()
dates = set()
researchers = set()
series = set()
for name in table["new_filename_iteration4_manualRename"].values:
    if not isinstance(name, str):
        assert np.isnan(name)
        continue
    if name == "delete":
        continue
    if "DROPFILE" in name:
        continue
    parts = name.split("_")
    if len(parts) < 5:
        print(f"Short name: {name}")
        continue
    if len(parts) > 6:
        print(f"long name: {name}")
    ids.add(parts[0])
    fixes.add(parts[1])
    for channel in parts[2].split("-"):
        channels.add(channel)
    dates.add(parts[3])
    researchers.add(parts[4])
    if len(parts) >5:
        series.add("_".join(parts[5:]))
    
```

```python
import subprocess
import sys


def get_n_channels_in_zarr_file(file_name:str, series:int):
    p1 = subprocess.check_output(
        [
            "mc",
            "ls",
            f"culcol_s3_ro/culture-collections/data/single_volumes/images/ome-zarr/{file_name}.ome.zarr/{series}/0/0",
        ]
    )
    n_channels_in_zarr_file = p1.decode(sys.stdout.encoding).count("\n")
    return n_channels_in_zarr_file
```

```python
# from tqdm import tqdm

mismatches = []
# for i, row in tqdm(table.iterrows(), total=table.shape[0]):
for i, row in table.iterrows():
    old_name = row["Filename"]
    new_name = row["new_filename_iteration4_manualRename"]
    if not isinstance(new_name, str):
        assert np.isnan(new_name)
        continue
    if new_name == "delete":
        continue
    if "DROPFILE" in new_name:
        continue
    split = new_name.split("_")
    series = 0
    if split[-1].startswith("s") and len(split[-1]) == 3:
        series = int(split[-1][1:])
        old_name = old_name.split("_")
        old_name = "_".join(old_name[:-1])
    n_channels_file = get_n_channels_in_zarr_file(old_name, series)
    channels = new_name.split("_")[2]
    n_channels_name = len(channels.split("-"))
    if n_channels_file != n_channels_name:
        print(f"{old_name} {new_name} {n_channels_file} {n_channels_name}")
        mismatches.append(
            {
                "index": i,
                "old_name": old_name,
                "new_name": new_name,
                "n_channels_file": n_channels_file,
                "n_channels_name": n_channels_name,
            }
        )
```
