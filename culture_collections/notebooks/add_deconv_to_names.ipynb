```python
c = 0
files_to_fix = []
for f in os.listdir(p):
    meta = p + f + "/OME/METADATA.ome.xml"
    tree = ET.parse(meta)
    root = tree.getroot()
    if "Lng" in ET.tostring(root, encoding="utf8").decode("utf8"):
        has_partners = False
        imgs = []
        img_names = []
        for el in root.iter():
            if el.tag.endswith("Image"):
                imgs.append(el)
                img_names.append(el.attrib["Name"])
        for img_name in img_names:
            if "_Lng" in img_name and img_name.split("_Lng")[0] in img_names:
                has_partners = True
                break
        if has_partners:
            c += 1
            print(f)
            files_to_fix.append(f)
c
```

```python
name_map = {}
old_names = []
new_names = []
for f in files_to_fix:
    print(f)
    view_name_base = f.replace(".ome.zarr", "")
    old_series_counter = 0
    new_series_counter = 0
    meta = p + f + "/OME/METADATA.ome.xml"
    tree = ET.parse(meta)
    root = tree.getroot()
    img_names = []
    for el in root.iter():
        if el.tag.endswith("Image"):
            img_names.append(el.attrib["Name"])
    print(len(img_names), "images found")
    for i, img_name in enumerate(img_names):
        old_view_name = view_name_base + f"_s{old_series_counter:02}"
        old_series_counter += 1

        if (  # has partner case
            "_Lng" in img_name and img_name.split("_Lng")[0] in img_names
        ):
            assert (  # make sure the partner is the one before
                    img_names[i - 1].split("_Lng")[0] == img_name.split("_Lng")[0]
                ), f"Partner mismatch for {img_name} in {f}"
            # partner is before and increased series counter
            new_view_name = view_name_base + f"_s{new_series_counter-1:02}"
        else:  # no partner case
            new_view_name = view_name_base + f"_s{new_series_counter:02}"
            new_series_counter += 1
        if "_Lng" in img_name:
            new_view_name += "_deconv"
        if "Align" in img_name:
            new_view_name += "_align"
        old_names.append(old_view_name)
        new_names.append(new_view_name)
        name_map[old_view_name] = new_view_name
        print(f"{old_view_name} -> {new_view_name} for {img_name}")
    print()
```
