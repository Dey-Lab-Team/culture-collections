```python
import re
import pandas as pd

table = pd.read_csv("final_name_map.csv", sep="\t")
table
```

```python
for name in table["new_filename_iteration4_manualRename"].tolist():
    if not isinstance(name, str):
        print("")
        continue
    pattern = re.compile(r"_(0\d\d)")
    match = pattern.search(name)
    if match:
        number = int(match.group(1))
        name = name.replace(match.group(1), f"{number:02}")
        print(name)
    else:
        print(name)
```
